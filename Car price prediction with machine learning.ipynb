Convert categorical data type to numerical data type

```python
#manual encoding
car_data.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)
#one hot encoding
car_data = pd.get_dummies(car_data, columns=['Selling_type', 'Transmission'], drop_first=True)
```

```python
car_data.head()
```

Heatmap

```python
# Assuming 'car_data' is your DataFrame
numeric_car_data = car_data.select_dtypes(include=np.number)

correlation_matrix_spearman = numeric_car_data.corr(method='spearman')

plt.figure(figsize=(10, 7))
sns.heatmap(data=correlation_matrix_spearman, annot=True)
plt.title('Spearman Correlation between the columns')
plt.show()
```

```python
car_data_numeric = car_data.apply(pd.to_numeric, errors='coerce')

correlation_matrix_spearman = car_data_numeric.corr(method='spearman')

plt.figure(figsize=(10, 7))
sns.heatmap(data=correlation_matrix_spearman, annot=True)
plt.title('Spearman Correlation between the columns')
plt.show()
```

Data

```python
X = car_data.drop(['Car_Name','Selling_Price'], axis=1)
y = car_data['Selling_Price']
print("Shape of X is: ",X.shape)
print("Shape of y is: ", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
print("X_test shape:", X_test.shape)
print("X_train shape:", X_train.shape)
print("y_test shape: ", y_test.shape)
print("y_train shape:", y_train.shape)
```

```python
#scaling
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Model

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE: ", (metrics.mean_absolute_error(pred, y_test)))
print("MSE: ", (metrics.mean_squared_error(pred, y_test)))
print("R2 score: ", (metrics.r2_score(pred, y_test)))
```

Model plot

```python
sns.regplot(x=pred, y=y_test)
plt.xlabel("Predicted Price")
plt.ylabel('Actual Price')
plt.title("Actual vs predicted price")
plt.show()
```
